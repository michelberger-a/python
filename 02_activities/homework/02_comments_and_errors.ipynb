### 2. Write a block of code to check if the value is a string. If it is, print out a message saying it is valid. Otherwise, raise an exception with a message saying the value must be a string.

````python
value = 6

# Your code goes here
if type(value) == "str":
    print("It is a valid string")
else:
    print("value needs to be converted to a string")

## or 

if type(value) == str:
    print("It is a valid string")
else:
    raise TypeError("value needs to be converted to a string")
````
